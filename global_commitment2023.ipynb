```python
# gcreport2023data_cpg = gcreport2023data_cpgsheet.rename


import pandas as pd

# Load the Excel file
file_path = '/mnt/data/Global Commitment Progress Report 2023 - Data Sheet (1).xlsx'
excel_data = pd.ExcelFile(file_path)

# Load all sheets into a dictionary of DataFrames
data_frames = {sheet: excel_data.parse(sheet) for sheet in excel_data.sheet_names}

# Function to remove spaces from column names
def remove_spaces_from_column_names(df):
    df.columns = df.columns.str.replace(' ', '')
    return df

# Apply the function to each DataFrame
modified_data_frames = {sheet: remove_spaces_from_column_names(df) for sheet, df in data_frames.items()}

# Save the modified DataFrames back to a new Excel file
modified_file_path = '/mnt/data/Global_Commitment_Progress_Report_2023_Modified.xlsx'
with pd.ExcelWriter(modified_file_path) as writer:
    for sheet, df in modified_data_frames.items():
        df.to_excel(writer, sheet_name=sheet, index=False)

# Path to the modified file
modified_file_path

```
